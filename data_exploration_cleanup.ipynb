```python
# Import dependancies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Create variables for raw data files
happiness_2015_file = 'Resources/2015.csv'
happiness_2016_file = 'Resources/2016.csv'
happiness_2017_file = 'Resources/2017.csv'
happiness_2018_file = 'Resources/2018.csv'
happiness_2019_file = 'Resources/2019.csv'
```

```python
# Read country data into notebook
footprint_data1 = pd.read_csv("Resources/countries.csv")
footprint_data1.head(1)
```

```python
#Change names of Countries to match the Country names in the happiness data
replacements = {'Country': {
r'Libyan Arab Jamahiriya':'Libya',
r'Macedonia TFYR':'Macedonia',
r'Russian Federation':'Russia',
r'Syrian Arab Republic':'Syria',
r'Tanzania, United Republic of':'Tanzania',
r'United States of America':'United States',
r'Venezuela, Bolivarian Republic of':'Venezuela'}}

footprint_data1.replace(replacements, regex=True, inplace=True)

footprint_data1.head()
```

```python
#Read Happiness data into notebook
Happiness_data1 = pd.read_csv("Resources/2016.csv")
Happiness_data1.head(1)
```

```python
#Merge Happiness and Footprint data
merge1 = footprint_data1.merge(Happiness_data1, left_on='Country', right_on='Country')
merge1
```

```python
# Drop the duplicate country column
merge2 = merge1.drop(merge1.columns[-1],1)
merge2.head(1)

# Drop columns that are not useful
merge2.drop(['Region_x','Cropland Footprint','Grazing Footprint','Forest Footprint','Carbon Footprint','Fish Footprint',
             'Cropland','Grazing Land','Forest Land','Fishing Water','Urban Land','Data Quality',
             'Lower Confidence Interval','Upper Confidence Interval'],axis=1, inplace=True)
merge2

# Export merged dataframe to csv
merge2.to_csv('Resources/countries_happiness_merge.csv')
```

```python
# Clean the 2015 data and insert the year
happiness_2015_df = pd.read_csv(happiness_2015_file)
happiness_2015_df.insert(0, 'Year', 2015)
del happiness_2015_df['Standard Error']
del happiness_2015_df['Dystopia Residual']
happiness_2015_df.rename(columns={'Family': 'Social Support'}, inplace=True)
happiness_2015_df                         
```

```python
# Clean the 2016 data and insert the year
happiness_2016_df = pd.read_csv(happiness_2016_file)
happiness_2016_df.insert(0, 'Year', 2016)
del happiness_2016_df['Lower Confidence Interval']
del happiness_2016_df['Upper Confidence Interval']
del happiness_2016_df['Dystopia Residual']
happiness_2016_df.rename(columns={'Family': 'Social Support'}, inplace=True)
happiness_2016_df
```

```python
# Clean the 2017 data and insert the year
happiness_2017_df = pd.read_csv(happiness_2017_file)
happiness_2017_df.insert(0, 'Year', 2017)
happiness_2017_df['Region'] = happiness_2017_df['Country'].map(happiness_2016_df.set_index('Country')['Region'])
region = happiness_2017_df['Region']
happiness_2017_df.drop(labels=['Region'], axis=1,inplace = True)
happiness_2017_df.insert(2, 'Region', region)
del happiness_2017_df['Whisker.high']
del happiness_2017_df['Whisker.low']
del happiness_2017_df['Dystopia.Residual']
happiness_2017_df.rename(columns={'Happiness.Rank': 'Happiness Rank', 
                                  'Happiness.Score': 'Happiness Score',
                                  'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
                                  'Family': 'Social Support',
                                  'Health..Life.Expectancy.': 'Health (Life Expectancy)', 
                                  'Trust..Government.Corruption.': 'Trust (Government Corruption)'}, inplace=True)
trust = happiness_2017_df['Trust (Government Corruption)']
happiness_2017_df.drop(labels=['Trust (Government Corruption)'], axis=1,inplace = True)
happiness_2017_df.insert(9, 'Trust (Government Corruption)', trust)
happiness_2017_df.head()
```

```python
# Clean the 2018 data and insert the year
happiness_2018_df = pd.read_csv(happiness_2018_file)
happiness_2018_df.insert(0, 'Year', 2018)
happiness_2018_df['Region'] = happiness_2018_df['Country or region'].map(happiness_2017_df.set_index('Country')['Region'])
region = happiness_2018_df['Region']
happiness_2018_df.drop(labels=['Region'], axis=1,inplace = True)
happiness_2018_df.insert(3, 'Region', region)
rank = happiness_2018_df['Overall rank']
happiness_2018_df.drop(labels=['Overall rank'], axis=1,inplace = True)
happiness_2018_df.insert(3, 'Happiness Rank', rank)
happiness_2018_df.rename(columns={'Country or region': 'Country',
                                  'Score': 'Happiness Score',
                                  'GDP per capita': 'Economy (GDP per Capita)',
                                  'Social support': 'Social Support',
                                  'Healthy life expectancy': 'Health (Life Expectancy)',
                                  'Freedom to make life choices': 'Freedom',
                                  'Perceptions of corruption': 'Trust (Government Corruption)'}, inplace=True)
trust = happiness_2018_df['Trust (Government Corruption)']
happiness_2018_df.drop(labels=['Trust (Government Corruption)'], axis=1,inplace = True)
happiness_2018_df.insert(9, 'Trust (Government Corruption)', trust)
happiness_2018_df
```

```python
# Clean the 2019 data and insert the year
happiness_2019_df = pd.read_csv(happiness_2019_file)
happiness_2019_df.insert(0, 'Year', 2019)
happiness_2019_df['Region'] = happiness_2019_df['Country or region'].map(happiness_2017_df.set_index('Country')['Region'])
region = happiness_2019_df['Region']
happiness_2019_df.drop(labels=['Region'], axis=1,inplace = True)
happiness_2019_df.insert(3, 'Region', region)
rank = happiness_2019_df['Overall rank']
happiness_2019_df.drop(labels=['Overall rank'], axis=1,inplace = True)
happiness_2019_df.insert(3, 'Happiness Rank', rank)
happiness_2019_df.rename(columns={'Country or region': 'Country',
                                  'Score': 'Happiness Score',
                                  'GDP per capita': 'Economy (GDP per Capita)',
                                  'Social support': 'Social Support',
                                  'Healthy life expectancy': 'Health (Life Expectancy)',
                                  'Freedom to make life choices': 'Freedom',
                                  'Perceptions of corruption': 'Trust (Government Corruption)'}, inplace=True)
trust = happiness_2019_df['Trust (Government Corruption)']
happiness_2019_df.drop(labels=['Trust (Government Corruption)'], axis=1,inplace = True)
happiness_2019_df.insert(9, 'Trust (Government Corruption)', trust)
happiness_2019_df
```

```python
# Combine all the year dataframes into a single dataframe
happiness_df_list = [happiness_2015_df, happiness_2016_df, happiness_2017_df, happiness_2018_df, happiness_2019_df]
happiness_df = pd.concat(happiness_df_list)
happiness_df

# Export combined dataframe to a csv
happiness_df.to_csv('Resources/happiness_years.csv', encoding='utf-8', index=False)
```

```python
# Create dataframes for top 10 ranked countries
top_happiness_2015 = happiness_df.loc[happiness_df['Year'] == 2015, :].head(10)
top_happiness_2015
top_happiness_2016 = happiness_df.loc[happiness_df['Year'] == 2016, :].head(10)
top_happiness_2016
top_happiness_2017 = happiness_df.loc[happiness_df['Year'] == 2017, :].head(10)
top_happiness_2017
top_happiness_2018 = happiness_df.loc[happiness_df['Year'] == 2018, :].head(10)
top_happiness_2018
top_happiness_2019 = happiness_df.loc[happiness_df['Year'] == 2019, :].head(10)
top_happiness_2019

top_10 = [top_happiness_2015, top_happiness_2016, top_happiness_2017, top_happiness_2018, top_happiness_2019]

top_10_df = pd.concat(top_10)
top_10_df

top_10_df.to_csv('Resources/happiness_top_10.csv', encoding='utf-8', index=False)
```

```python
bottom_happiness_2015 = happiness_df.loc[happiness_df['Year'] == 2015, :].tail(10)
bottom_happiness_2015
bottom_happiness_2016 = happiness_df.loc[happiness_df['Year'] == 2016, :].tail(10)
bottom_happiness_2016
bottom_happiness_2017 = happiness_df.loc[happiness_df['Year'] == 2017, :].tail(10)
bottom_happiness_2017
bottom_happiness_2018 = happiness_df.loc[happiness_df['Year'] == 2018, :].tail(10)
bottom_happiness_2018
bottom_happiness_2019 = happiness_df.loc[happiness_df['Year'] == 2019, :].tail(10)
bottom_happiness_2019

bottom_10 = [bottom_happiness_2015, bottom_happiness_2016, bottom_happiness_2017, bottom_happiness_2018, 
             bottom_happiness_2019]

bottom_10_df = pd.concat(bottom_10)
bottom_10_df

bottom_10_df.to_csv('Resources/happiness_bottom_10.csv', encoding='utf-8', index=False)
```

```python
top_bottom_df = top_10_df.append(bottom_10_df)
top_bottom_df

top_bottom_df.to_csv('Resources/happiness_top_bottom_10.csv', encoding='utf-8', index=False)
```
