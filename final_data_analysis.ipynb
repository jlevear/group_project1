```python
# Import dependancies and locate/read csv file
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

happiness_file = 'Resources/happiness_years.csv'
top_10_file = 'Resources/happiness_top_10.csv'
bottom_10_file = 'Resources/happiness_bottom_10.csv'
top_bottom_10_file = 'Resources/happiness_top_bottom_10.csv'
region_avgs_file = 'Resources/happiness_region_avgs.csv'
us_file = 'Resources/happiness_us.csv'
merge_file = 'Resources/countries_happiness_merge.csv'

happiness_df = pd.read_csv(happiness_file)
top_10_df = pd.read_csv(top_10_file)
bottom_10_df = pd.read_csv(bottom_10_file)
top_bottom_10_df = pd.read_csv(top_bottom_10_file)
region_avgs_df = pd.read_csv(region_avgs_file)
us_df = pd.read_csv(us_file)
merge_df = pd.read_csv(merge_file)
```

```python
# Happiness score vs GDP 
x_axis1 = happiness_df['Happiness Rank']
y_axis1 = happiness_df['Economy (GDP per Capita)']

x_axis2 = top_bottom_10_df['Happiness Rank']
y_axis2 = top_bottom_10_df['Economy (GDP per Capita)']

x_axis3 = us_df['Happiness Rank']
y_axis3 = us_df['Economy (GDP per Capita)']

plt.figure(figsize=(12,6))

# (code from url: 
# https://stackoverflow.com/questions/20105364/how-can-i-make-a-scatter-plot-colored-by-density-in-matplotlib)
# (User: Joe Kington)
xy = np.vstack([x_axis1,y_axis1])
z = gaussian_kde(xy)(xy)

plt.scatter(x_axis1, y_axis1, c=z, s=50, edgecolor='')

plt.scatter(x_axis2, y_axis2, marker='o', facecolors='tab:blue', edgecolors='red', alpha=0.75)

plt.scatter(x_axis3, y_axis3, marker='*', facecolors='black', s=100)

plt.ylabel('Economy (GDP per Capita)')
plt.xlabel('Happiness Rank')
plt.title('Country Happiness vs GDP over 5 Years')
plt.grid(b=True)
plt.gca().invert_xaxis()


plt.savefig('Images/Country Happiness vs GDP over 5 Years.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
regroup_country = merge_df.groupby(["Country"])
regroup_country
```

```python
us = regroup_country.groups["United States"]
us = regroup_country.get_group("United States")
us
```

```python
# where is the US? happiness rank and biocapacity deficit/reserve
x_axis1 = us['Happiness Rank']
y_axis1 = us['Biocapacity Deficit or Reserve']

x_axis2 = merge_df['Happiness Rank']
y_axis2 = merge_df['Biocapacity Deficit or Reserve']

plt.figure(figsize=(12,6))

# (code from url: 
# https://stackoverflow.com/questions/20105364/how-can-i-make-a-scatter-plot-colored-by-density-in-matplotlib)
# (User: Joe Kington)
xy = np.vstack([x_axis2,y_axis2])
z = gaussian_kde(xy)(xy)

plt.scatter(x_axis2, y_axis2, c=z, s=100, edgecolor='')

plt.scatter(x_axis1, y_axis1, marker='*', facecolors='tab:red', s=300, alpha=0.75)

plt.ylabel('Biocapacity Deficit or Reserve')
plt.xlabel('Happiness Rank')
plt.title('Biocapacity Deficit or Reserve vs Country Happiness')
plt.grid(b=True)
plt.gca().invert_xaxis()
plt.ylim(top=3, bottom=-5)

plt.savefig('Images/Biocapacity Deficit or Reserve vs Country Happiness.png', dpi=300, bbox_inches='tight')
plt.show()
```
